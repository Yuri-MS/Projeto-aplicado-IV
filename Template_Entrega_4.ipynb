# **Referencial Teórico**

1 Técnicas de Análise de Séries Temporais
Os métodos de previsão utilizam dados passados para estimar valores
futuros. Até o início do século XX, as previsões eram feitas com simples
extrapolações temporais. YULE (1927) trouxe uma inovação ao propor que séries
temporais fossem tratadas como processos estocásticos, resultando no modelo
autorregressivo (AR), baseado em valores anteriores.
Apesar dos avanços, modelos lineares apresentavam limitações para capturar
a não linearidade de muitas séries, como apontado por CHAVES (1991). A técnica
de alisamento exponencial, surgida na década de 1950, ajudou a melhorar as
previsões com o uso de computadores.
BOX e JENKINS (1976) consolidaram o conhecimento sobre séries
temporais, focando na parcimônia dos modelos estocásticos. Mais recentemente,GOOIJER e HYNDMAN (2006) revisaram os principais estudos sobre o tema nos
últimos 25 anos.

2 Modelo ARIMA
O modelo ARIMA (AutoRegressive Integrated Moving Average) é uma técnica
amplamente utilizada para previsão de séries temporais, combinando três
componentes: auto regressão (AR), integração (I) e média móvel (MA). O
componente AR utiliza valores passados da série para prever valores futuros, a
integração (I) transforma séries não estacionárias em estacionárias através da
diferenciação, e o componente MA ajusta as previsões atuais com base nos erros de
previsão passados.
Representado como ARIMA (p, d, q), onde ppp é a ordem da auto regressão, ddd é
o número de diferenciações necessárias para tornar a série estacionária e qqq é a
ordem da média móvel, o modelo é ajustado observando padrões de autocorrelação.
Após a seleção dos parâmetros, o ARIMA pode ser usado para prever valores
futuros, com base nos padrões históricos da série temporal.
Aplicado em áreas como finanças, economia e previsão de eventos naturais, o
ARIMA é útil para identificar tendências temporais. No caso da previsão de
queimadas, ele pode capturar padrões históricos e sazonais, fornecendo uma base
para estimar futuras ocorrências, especialmente se forem observadas tendências
repetitivas nos dados.

3 Modelo SARIMA
O modelo SARIMA (Seasonal AutoRegressive Integrated Moving Average) é
uma extensão do ARIMA que inclui a sazonalidade nas séries temporais. Ele é
representado como SARIMA (p, d, q)(P, D, Q)[S], onde p, d, e q se referem aos
componentes não sazonais, enquanto P, D, e Q são os componentes sazonais, e S
indica a periodicidade da sazonalidade, como 12 para dados mensais.
A principal vantagem do SARIMA é sua capacidade de capturar padrões
sazonais e cíclicos, essenciais para previsões em contextos como vendas, clima e
fenômenos naturais. Para ajustar o modelo, é necessário garantir que a série seja
estacionária, o que pode envolver diferenciação para componentes sazonais e não
sazonais. A análise de autocorrelação (ACF) e autocorrelação parcial (PACF) ajuda
na seleção dos parâmetros apropriados.
Uma vez ajustado, o SARIMA pode fornecer previsões mais precisas para
séries com comportamento sazonal. Por exemplo, na previsão de queimadas, o
modelo pode identificar padrões sazonais, como a ocorrência de incêndios em
determinadas épocas do ano, melhorando a precisão das estimativas futuras. Essa
combinação de tendências de longo prazo e flutuações sazonais torna o SARIMA
uma ferramenta valiosa em diversas áreas.

4 Redes Neurais Recorrentes
Redes neurais recorrentes (RNNs) são ideais para análise de séries
temporais, pois podem capturar dependências entre dados passados e futuros.
Diferente de redes tradicionais, as RNNs têm uma memória interna que permite
modelar relações sequenciais, sendo úteis em previsões como clima e finanças.
Uma variante popular é a Long Short-Term Memory (LSTM), que resolve
problemas em sequências longas, mantendo informações relevantes por mais
tempo. Isso as torna eficazes para capturar padrões complexos e de longo prazo,
como em flutuações de mercado ou demanda energética.
RNNs são vantajosas para séries temporais porque, ao contrário de modelos
lineares, como ARIMA, elas aprendem padrões não lineares diretamente dos dados.
Apesar de exigirem mais recursos para treinamento, oferecem previsões mais
precisas em cenários complexos.
